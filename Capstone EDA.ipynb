```python
clean_2 = clean_1[['Time', 'Country Name','Country Code','GNI, Atlas method (current US$) [NY.GNP.ATLS.CD]' ,'GNI per capita, Atlas method (current US$) [NY.GNP.PCAP.CD]' ,'Surface area (sq. km) [AG.SRF.TOTL.K2]','Population, total [SP.POP.TOTL]', 'Population growth (annual %) [SP.POP.GROW]', 'GDP per capita (current US$) [NY.GDP.PCAP.CD]']]
clean_2
```

```python
clean_2.info()
```

```python
clean_2 = clean_2.rename(columns={"Time": "year", "Country Name": "country_name","GNI, Atlas method (current US$) [NY.GNP.ATLS.CD]":"GNI","GNI per capita, Atlas method (current US$) [NY.GNP.PCAP.CD]":"GNI_per_capita", "Country Code":"country_code", "Surface area (sq. km) [AG.SRF.TOTL.K2]":"surface_area(sqkm)", "Population, total [SP.POP.TOTL]":"total_population", "Population growth (annual %) [SP.POP.GROW]":"population_growth", "GDP per capita (current US$) [NY.GDP.PCAP.CD]":"GDP_per_capita"})
clean_2.head()
```

```python
clean_2['year'] = clean_2['year'].astype(int)
```

```python
clean_2['country_name'].unique()
```

```python
clean_2['country_code'].unique()
```

```python
clean_5 =  """SELECT country_name, year , country_code , total_population, population_growth, GDP_per_capita, GNI, GNI_per_capita FROM clean_2 WHERE  year >=2000  ORDER BY year"""
clean_5_pop1 = ps.sqldf(clean_5)
clean_5_pop1
```
